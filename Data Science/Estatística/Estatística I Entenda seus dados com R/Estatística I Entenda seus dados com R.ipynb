# Função própria

```r
mode <- function(x) {
     ux <- unique(x)
     ux[which.max(tabulate(match(x, ux)))]
}
```

```r
lista_mode = c(1, 2, 2, 2, 2, 3, 3, 4, 5, 5, 6, 7)
mode(lista_mode)
```
